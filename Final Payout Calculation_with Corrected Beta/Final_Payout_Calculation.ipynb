```python
import numpy as np

# Inputs (December 2025)
ST_rss1 = 720.0          # Example: actual Dec 2025 price for RSS 1
ST_rss3 = 780.0          # Example: actual Dec 2025 price for RSS 3
realized_beta = 1.25     # Fuzzy output from actual 2025 rainfall + rainy days
capped_beta = min(realized_beta, 1.50)

# Strike prices (same as premium calculation)
strikes = {
    '100% Market': 700.0,
    '95% Market': 665.0,
    'Historical Mean': 424.77,
    'Production Cost': 386.08
}

print("FINAL PAYOUT (LKR per kg) - December 2025")
for label, K in strikes.items():
    payout_rss1 = max(K - ST_rss1, 0) * capped_beta
    payout_rss3 = max(K - ST_rss3, 0) * capped_beta
    print(f"{label:15} → RSS1: {payout_rss1:.2f}   RSS3: {payout_rss3:.2f}")
```
