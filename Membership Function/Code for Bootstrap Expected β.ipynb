```python
import numpy as np

# ================== YOUR DATA GOES HERE ==================
# Example: Replace with your 12 years of data
historical_rainfall = np.array([1850, 2650, 1420, 3120, 1980, 2750, 920, 3450, 2210, 1580, 2890, 3100])   # your actual mm values
historical_rainy_days = np.array([165, 178, 82, 215, 172, 195, 67, 232, 181, 145, 203, 218])             # your actual days values

# Simple fuzzy function to compute β for one year (you can expand this)
def compute_beta(rain, days):
    # Very simplified version of your FIS for demonstration
    if rain < 1750 and days < 100:
        return 1.38
    elif 2000 <= rain <= 3000 and 150 <= days <= 200:
        return 1.00
    elif rain > 3000 and days > 200:
        return 1.50          # capped
    else:
        return 1.18          # medium case

# ================== BOOTSTRAP TO GET EXPECTED β ==================
N = 10000
bootstrap_betas = []

for _ in range(N):
    # Resample one year randomly with replacement
    idx = np.random.randint(0, len(historical_rainfall))
    rain_sample = historical_rainfall[idx]
    days_sample = historical_rainy_days[idx]
    
    beta_sample = compute_beta(rain_sample, days_sample)
    bootstrap_betas.append(beta_sample)

expected_beta = np.mean(bootstrap_betas)
print(f"Expected β from Bootstrap (10,000 simulations) = {expected_beta:.2f}")
```
