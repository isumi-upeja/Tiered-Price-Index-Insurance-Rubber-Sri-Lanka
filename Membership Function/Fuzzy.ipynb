```python
import matplotlib.pyplot as plt
import numpy as np

def trap_mf(x, a, b, c, d):
    return np.maximum(0, np.minimum(np.minimum((x - a)/(b - a), 1), (d - x)/(d - c)))

# ================== FIGURE 3.1 ==================
plt.figure(figsize=(8, 5))
x_rain = np.linspace(0, 5000, 600)
plt.plot(x_rain, trap_mf(x_rain, 500, 1200, 1700, 1850), color='#1f77b4', lw=2.5, label='Low (<1750 mm)')
plt.plot(x_rain, trap_mf(x_rain, 1900, 2100, 2900, 3100), color='#2ca02c', lw=2.5, label='Medium (2000–3000 mm)')
plt.plot(x_rain, trap_mf(x_rain, 2950, 3200, 3800, 4500), color='#d62728', lw=2.5, label='High (>3000 mm)')
plt.title('Figure 3.1: Membership function for Annual Rainfall (mm/year)')
plt.xlabel('Annual Rainfall (mm)')
plt.ylabel('Degree of membership')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figure_3_1.png', dpi=400, bbox_inches='tight')
print("✅ Saved: figure_3_1.png")

# ================== FIGURE 3.2 ==================
plt.figure(figsize=(8, 5))
x_days = np.linspace(0, 260, 600)
plt.plot(x_days, trap_mf(x_days, 0, 40, 80, 95), color='#1f77b4', lw=2.5, label='Low (0–90 days)')
plt.plot(x_days, trap_mf(x_days, 145, 160, 195, 205), color='#2ca02c', lw=2.5, label='Medium (150–200 days)')
plt.plot(x_days, trap_mf(x_days, 195, 215, 235, 260), color='#d62728', lw=2.5, label='High (180–240+ days)')
plt.title('Figure 3.2: Membership function for Rain Interference Days')
plt.xlabel('Rain Interference Days (days/year)')
plt.ylabel('Degree of membership')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figure_3_2.png', dpi=400, bbox_inches='tight')
print("✅ Saved: figure_3_2.png")

# ================== FIGURE 3.3 ==================
plt.figure(figsize=(8, 5))
x_beta = np.linspace(0.95, 1.60, 500)
plt.plot(x_beta, trap_mf(x_beta, 0.98, 1.00, 1.08, 1.12), color='#1f77b4', lw=2.5, label='Very Low Risk')
plt.plot(x_beta, trap_mf(x_beta, 1.10, 1.15, 1.25, 1.30), color='#2ca02c', lw=2.5, label='Medium Risk')
plt.plot(x_beta, trap_mf(x_beta, 1.28, 1.35, 1.45, 1.55), color='#d62728', lw=2.5, label='Very High Risk (capped at 1.50)')
plt.title('Figure 3.3: Membership function for Weather Risk Loading Factor (β)')
plt.xlabel('Risk Loading Factor (β)')
plt.ylabel('Degree of membership')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figure_3_3.png', dpi=400, bbox_inches='tight')
print("✅ Saved: figure_3_3.png")

print("\n🎉 All three figures saved separately as PNG files!")
```
